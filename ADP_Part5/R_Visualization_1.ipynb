## 가. 기본 패키지 설치

```r
install.packages('ggplot2')
library(ggplot2)
```

## 나. XY 그래프

### 1) line 그래프

```r
data(ChickWeight)
head(ChickWeight)
```

```r
# aes : x, y, color, group
ggplot(ChickWeight, aes(x = Time, y = weight, color = Diet, group=Chick)) + geom_line()
# 먹이별 체중 변화가 나타나고 있지만, 어느 먹이(diet)가 효율적인지는 알기 어렵다.
```

### 2) 포인트 그래프

```r
h = ggplot(ChickWeight, aes(x=Time, y=weight, color = Diet ))
# alpha = 0 ~ 1 사이의 값 : 투명도
h + geom_point(alpha=0.5)
```

### 3) 스무스 그래프

```r
h = ggplot(ChickWeight, aes(x=Time, y=weight, color = Diet))
h+geom_smooth(alpha = 0.5, size = 1)
```

### 4) 개선된 그래프

```r
ggplot(ChickWeight, aes(x=Time, y=weight, color=Diet)) + geom_point(alpha=0.5) + geom_smooth(alpha=0.5, size=1)
```

## 다. 히스토그램

```r
ggplot(subset(ChickWeight, Time=21), aes(x=weight, fill=Diet)) + geom_histogram(color = 'black', binwidth=50) + facet_grid(. ~ Diet )
```

```r
ggplot(subset(ChickWeight, Time=21), aes(x=weight, fill=Diet)) + geom_histogram(color = 'black', binwidth=50) + facet_grid(Diet ~ . )
```

## 라. 포인트 그래프

```r
data(mtcars)
head(mtcars)
```

### 2) 기본 포인트 그래프

```r
p = qplot(wt, mpg, color = hp, data = mtcars)
p + coord_cartesian(ylim = c(0,40))
p + scale_color_continuous(breaks = c(100,300))
p + guides(color = 'colorbar')
```

## 사. 효과주기

### 1) 기본 효과 주기의 예시

### 4) 기타 효과 옵션의 예시

#### 가) 선형 모델링

```r
dmod = lm(price ~ cut, data = diamonds)
cuts = data.frame(cut = unique(diamonds$cut), predict(dmod, data.frame(cut=unique(diamonds$cut)), se=TRUE)[c('fit', 'se.fit')])
se = ggplot(cuts, aes(x =cut, y = fit, ymin = fit - se.fit, ymax = fit + se.fit, color = cut))
se + geom_pointrange()
```

#### 나) 포인트 그래프 - 박스로 강조

```r
p = ggplot(mtcars, aes(wt, mpg)) + geom_point()
p + annotate('rect', xmin = 2, xmax = 3.5, ymin = 2, ymax = 25, fill = 'dark grey', alpha = 0.5)
```

#### 다) 축의 범위 지정

```r
p = qplot(disp, wt, data = mtcars) + geom_smooth()
p + scale_x_continuous(limits = c(325, 500))
```
